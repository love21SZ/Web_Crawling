```python
import re


p =re.compile("ca.e")
# . (ca.e) : 하나의 문자를 의미 > care, cafe, case | caffe (X)
# ^ (^de)  : 문자열의 시작 > desk, destination (O) | fade (X)
# $ (se$)  : 문자열의 끝 > case, base (O) | face(X)

# print(m.group())

def print_match(m):
    if m:
        print(m.group())
    else:
        print("매칭되지 않음")
    
m = p.match("case")
print_match(m)
```

```python
def print_match(m):
    if m:
        print("m.group():", m.group()) # 일치하는 문자열 반환
        print("m.string:", m.string)
        print("m.start():", m.start())
        print("m.end():", m.end())
        print("m.span():", m.span())
    else:
        print("매칭되지 않음")
    
m = p.match("careless")
print_match(m)
```

```python
m = p.search("good care")
print_match(m)
```
